```python
from sklearn import datasets
import pandas as pd
```

```python
books_dataset = pd.read_csv('books_dataset_v5_titlefixed_titleinfo.csv')
```

```python
books_dataset
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = books_dataset[['Number of Pages', 'Subject Place', 'Subject', 'publisher_id', 'publishplace_id', 'dewey_id', 'publisher_freq', 'publish_place_freq', 'dewey_freq', 'author_id', 'author_frequency', 'title_avg_word_len', 'title_num_words']]
#removed publish date
```

```python
X
```

```python
y = books_dataset['on_NYT']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=251)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 100)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics

print('Accuracy:', metrics.accuracy_score(y_test, y_pred))
```

```python
feature_imp = pd.Series(clf.feature_importances_, index=['Number of Pages', 'Subject Place', 'Subject', 'publisher_id', 'publishplace_id', 'dewey_id', 'publisher_freq', 'publish_place_freq', 'dewey_freq', 'author_id', 'author_frequency', 'title_avg_word_len', 'title_num_words' ]).sort_values(ascending=False)
```

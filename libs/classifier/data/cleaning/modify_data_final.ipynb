```python
import pandas as pd
import numpy as np
```

```python
books_ds = pd.read_csv('books_dataset_final_fixedtitle_filled_meanpages_2005pubdate.csv', index_col='ISBN')
```

```python
books_ds.head()
```

```python
formatted_title_numwords = []
formatted_title_avgwordlen = []
for x in books_ds['Title']:
    linearr = str(x).strip().split()
    numwords = len(linearr)
    formatted_title_numwords.append(numwords)
    sumw = 0
    for w in linearr:
        sumw += len(w)
    formatted_title_avgwordlen.append(sumw/numwords)
```

```python
formatted_title_numwords
```

```python
formatted_title_avgwordlen
```

```python
books_ds['title_avg_word_len'] = formatted_title_avgwordlen
books_ds['title_num_words'] = formatted_title_numwords
```

```python
books_ds.to_csv('books_dataset_v5.csv')
```

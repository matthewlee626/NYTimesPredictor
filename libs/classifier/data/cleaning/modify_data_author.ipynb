```python
import pandas as pd
import numpy as np
```

```python
books_ds = pd.read_csv('books_dataset_v3.csv', index_col='ISBN')
```

```python
books_ds
```

```python
no = pd.read_csv('no.csv')
yes = pd.read_csv('yes.csv')
```

```python
yes
```

```python
formatted_authors = []

for i in range(no.shape[0]):
    formatted_authors.append(no['Author'][i])

for i in range(yes.shape[0]):
    formatted_authors.append(yes['Author'][i])
```

```python
formatted_authors
```

```python
len(formatted_authors)
```

```python
books_ds['Author'] = formatted_authors
```

```python
books_ds
```

```python
author_list = []
for x in books_ds['Author']:
    author_list.append(x)
unique_author_list = list(set(author_list))
```

```python
author_dict = {}
for i in range(len(unique_author_list)):
    author_dict[unique_author_list[i]] = i
```

```python
formatted_author = []
for x in books_ds['Author']:
    formatted_author.append(author_dict[x])
```

```python
books_ds['author_id'] = formatted_author
```

```python
books_ds
```

```python
from collections import Counter
author_count_dict = Counter(yes['Author'])
author_count_freq_list = []
for x in books_ds['Author']:
    if pd.isnull(x) or x == np.nan:
        author_count_freq_list.append(0)
    else:
        author_count_freq_list.append(author_count_dict[x])
```

```python
author_count_dict
```

```python
len(author_count_dict)
```

```python
len(unique_author_list)
```

```python
books_ds['author_frequency'] = author_count_freq_list
```

```python
books_ds
```

```python
books_ds.to_csv('books_dataset_v4.csv')
```

```python
publisherlist = []
for x in books_DS['Publisher']:
    publisherlist.append(x)
```

```python
len(publisherlist)
```

```python
publisherSet = set(publisherlist)
```

```python
len(publisherSet)
```

```python
unique_pub_list = list(publisherSet)
```

```python
dict = {}
for i in range(len(unique_pub_list)):
    dict[unique_pub_list[i]] = i
```
